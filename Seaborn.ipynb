# Pendahuluan Seaborn

## Apa itu Seaborn?
Seaborn adalah library visualisasi data berbasis Python yang dibangun di atas Matplotlib. Library ini dirancang untuk membuat visualisasi data statistik yang menarik dan mudah dipahami dengan sintaks yang lebih sederhana dibandingkan dengan Matplotlib. Seaborn menyediakan fungsi-fungsi tingkat tinggi untuk membuat plot kompleks dengan sedikit kode.

**Fitur Utama Seaborn:**
- Membuat plot statistik dengan mudah.
- Mendukung berbagai jenis visualisasi data seperti scatter plot, bar plot, box plot, dan heatmap.
- Otomatis melakukan agregasi data dan menambahkan elemen-elemen statistik.
- Mendukung penggunaan tema dan palet warna yang menarik.

---

## Perbandingan Seaborn dengan Matplotlib

| Aspek             | Matplotlib                         | Seaborn                                    |
|-------------------|------------------------------------|--------------------------------------------|
| Sintaks           | Lebih kompleks                     | Lebih sederhana                            |
| Visualisasi       | Visualisasi dasar                  | Visualisasi data statistik yang kompleks    |
| Palet Warna       | Perlu diatur manual                | Tersedia berbagai palet bawaan             |
| Integrasi         | Terbatas                           | Terintegrasi dengan Pandas                 |
| Statistik         | Perlu perhitungan manual           | Otomatis menghitung elemen statistik       |

Contoh kode dengan Matplotlib:
```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.show()
```

Contoh kode dengan Seaborn:
```python
import seaborn as sns
import pandas as pd

# Membuat DataFrame
data = pd.DataFrame({'x': [1, 2, 3, 4], 'y': [10, 20, 25, 30]})

# Membuat plot
sns.lineplot(data=data, x='x', y='y')
```

---

## Mengapa Menggunakan Seaborn?

1. **Visualisasi Lebih Menarik:**
   Seaborn secara otomatis menambahkan elemen visual seperti grid, warna, dan label yang membuat plot lebih menarik.

2. **Mudah Digunakan:**
   Dibandingkan dengan Matplotlib, Seaborn memiliki sintaks yang lebih sederhana dan ringkas.

3. **Integrasi dengan Pandas:**
   Seaborn bekerja sangat baik dengan DataFrame Pandas, memudahkan proses manipulasi dan visualisasi data.

4. **Plot Statistik:**
   Seaborn mempermudah pembuatan plot statistik seperti distribusi data, hubungan antar variabel, dan heatmap korelasi.

# Mengelola Palet Warna di Seaborn

Warna memainkan peran penting dalam visualisasi data, memperkenalkan berbagai dimensi pada plot jika digunakan dengan efektif. Anggaplah palet warna seperti kanvas yang digunakan seorang pelukis untuk mencampur cat.

## Fungsi `color_palette()`

**Seaborn** memperkenalkan fungsi yang sangat berguna bernama **`color_palette()`**. Fungsi ini memungkinkan kita menyisipkan warna-warna cerah ke dalam plot, sehingga meningkatkan daya tarik visualnya. Fungsi ini menghasilkan daftar warna yang bersama-sama membentuk sebuah palet warna.

## Palet Warna Bawaan di Seaborn

**Seaborn** menyediakan beberapa palet bawaan yang memiliki karakteristik unik masing-masing:

- **Deep**
- **Muted**
- **Bright**
- **Pastel**
- **Dark**
- **Colorblind**

Selain palet bawaan ini, kita juga dapat membuat **palet warna kustom** sesuai keinginan.

# 2. Distribution Plot

Distribusi data adalah hal yang paling penting untuk dipahami saat menganalisis data. Dalam bagian ini, kita akan melihat bagaimana Seaborn membantu kita memahami distribusi data kita.

## 2.1 histplot

Fungsi `histplot()` adalah cara paling mudah untuk melihat distribusi univariat dengan cepat. Fungsi ini akan membuat histogram yang sesuai dengan estimasi kepadatan kernel (Kernel Density Estimation atau KDE) dari data.


### Panduan Sintaks `sns.histplot`

#### 1. `sns.histplot` Options

Berikut adalah beberapa opsi atau parameter yang dapat digunakan dengan fungsi `sns.histplot`:

1. **`data`**: Dataset dari mana kolom yang ingin dipetakan diambil.
2. **`x`**: Nama kolom atau array untuk sumbu x.
3. **`y`**: Nama kolom atau array untuk sumbu y (jika membuat histogram 2D).
4. **`hue`**: Nama kolom untuk pemisahan warna.
5. **`bins`**: Jumlah bin dalam histogram.
6. **`binwidth`**: Lebar setiap bin.
7. **`binrange`**: Rentang nilai untuk bin.
8. **`stat`**: Menentukan statistik yang akan dihitung dalam setiap bin, seperti `'count'`, `'frequency'`, `'density'`, atau `'probability'`.
9. **`cumulative`**: Jika `True`, membuat histogram kumulatif.
10. **`element`**: Jenis elemen yang akan digunakan dalam plot, seperti `'bars'`, `'step'`, atau `'poly'`.
11. **`multiple`**: Bagaimana histogram harus digabungkan untuk variabel kategori, seperti `'layer'`, `'dodge'`, atau `'stack'`.
12. **`kde`**: Menambahkan plot KDE di atas histogram (`True` atau `False`).
13. **`color`**: Warna plot.
14. **`fill`**: Jika `True`, mengisi area di bawah histogram.

##### Contoh `sns.histplot` dengan beberapa opsi:

```python
sns.histplot(data=full_data, x='Age', bins=20, kde=True, color='blue', stat='density', cumulative=False)
plt.show()
```

Mari kita buat histogram dari jumlah orang tua dan anak yang ikut dalam perjalanan penumpang (kolom `Parch`).

```python
sns.histplot(full_data['Parch'], kde=False)
plt.show()
```

Dari grafik yang dihasilkan, kita dapat melihat bahwa sebagian besar penumpang tidak memiliki orang tua maupun anak yang ikut dalam perjalanan.

````python
sns.histplot(data=data, x='parch', kde=True, fill=True)
````

## 2.2 kdeplot

Kernel Density Estimation (KDE) adalah metode untuk memperkirakan fungsi kepadatan probabilitas dari variabel acak kontinu. Ini digunakan untuk analisis non-parametrik.


# Panduan Sintaks  `sns.kdeplot` dalam Seaborn

## 2. `sns.kdeplot` Options

Berikut adalah beberapa opsi atau parameter yang dapat digunakan dengan fungsi `sns.kdeplot`:

1. **`data`**: Dataset dari mana kolom yang ingin dipetakan diambil.
2. **`x`**: Nama kolom atau array untuk sumbu x.
3. **`y`**: Nama kolom atau array untuk sumbu y (untuk KDE 2D).
4. **`shade`**: Jika `True`, mengisi area di bawah kurva KDE.
5. **`bw_adjust`**: Mengatur bandwidth KDE. Nilai lebih besar memperhalus kurva.
6. **`cut`**: Memperpanjang plot di luar data.
7. **`clip`**: Rentang nilai untuk memotong kurva.
8. **`cumulative`**: Jika `True`, membuat KDE kumulatif.
9. **`hue`**: Memisahkan KDE berdasarkan kategori.
10. **`palette`**: Palet warna untuk variabel kategori.
11. **`fill`**: Mengisi area di bawah kurva KDE (seperti `shade`).
12. **`levels`**: Jumlah atau daftar level kontur untuk KDE 2D.

### Contoh `sns.kdeplot` dengan beberapa opsi:

```python
sns.kdeplot(data=full_data, x='Age', shade=True, bw_adjust=0.5, color='red', cumulative=False)
plt.show()
```

# Seaborn Relational Plots

Relational plots adalah jenis plot yang digunakan untuk menampilkan hubungan antara dua variabel. Seaborn menyediakan beberapa fungsi utama untuk membuat relational plots, seperti `relplot()`, `scatterplot()`, dan `lineplot()`. Berikut adalah penjelasan lengkap mengenai masing-masing fungsi tersebut.



## 3.1. lineplot

**Line plot** digunakan untuk menggambar garis yang menunjukkan hubungan antara dua variabel. Fungsi ini mendukung beberapa pengelompokan semantik menggunakan parameter seperti `hue`, `size`, dan `style`.

###  **Sintaks Dasar**:
```python
plt.figure(figsize=(8, 8))
sns.lineplot(x="Age", y="Fare", hue="Sex", style="Sex", data=full_data)
plt.show()
```

###  **Parameter Penting**:
- `x` dan `y`: Nama kolom dalam dataset yang digunakan sebagai sumbu horizontal dan vertikal.
- `hue`: Mengubah warna garis berdasarkan nilai kategori.
- `style`: Mengubah gaya garis berdasarkan kategori.
- `size`: Mengubah ketebalan garis berdasarkan nilai numerik.

###  **Fitur Tambahan**:
- Secara default, `lineplot` akan menghitung **estimasi kecenderungan sentral** dan **interval kepercayaan**.
- Parameter `ci` dapat digunakan untuk mengontrol interval kepercayaan:
  - `ci="sd"` akan menampilkan standar deviasi.
  - `ci=None` akan menghapus interval kepercayaan.

###  **Contoh Penggunaan**:
```python
# Line plot dengan interval kepercayaan
plt.figure(figsize=(8, 8))
sns.lineplot(x="Age", y="Fare", hue="Sex", style="Sex", ci=None, data=full_data)
plt.show()
```


````python
import seaborn as sns

data = sns.load_dataset('titanic')
data.head()
````

````python
sns.lineplot(data=data, x='age',hue='sex',style='sex', y='fare', errorbar=None)
````

# 4. **Plot Kategori (Categorical Plot) dalam Seaborn**

Dalam analisis data, jika salah satu atau kedua variabel yang dikaji bersifat kategori (categorical), maka kita dapat menggunakan plot kategori seperti `barplot()`, `stripplot()`, `swarmplot()`, dan lainnya yang disediakan oleh Seaborn.

## 4.1. **Barplot**

**Barplot** di Seaborn adalah salah satu visualisasi yang digunakan untuk menampilkan data kategori dengan nilai agregasi (seperti mean atau sum) untuk setiap kategori. Barplot memudahkan analisis perbandingan antar kategori dengan memvisualisasikan tinggi bar yang merepresentasikan nilai numerik.

Seaborn menyediakan **barplot** dengan sintaks sederhana dan banyak parameter yang bisa digunakan untuk menyesuaikan plot sesuai kebutuhan.

---

###  **Sintaks Dasar**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Membuat barplot sederhana
sns.barplot(x="kategori", y="nilai", data=df)
plt.show()
```

---

###  **Parameter dalam `sns.barplot()`**

| Parameter       | Deskripsi                                         | Default Value     |
|-----------------|---------------------------------------------------|-------------------|
| `x`             | Data kategori untuk sumbu X                       | -                 |
| `y`             | Data numerik untuk sumbu Y                        | -                 |
| `data`          | DataFrame yang berisi data                        | -                 |
| `hue`           | Variabel kategori tambahan untuk pembeda warna    | `None`            |
| `ci`            | Interval kepercayaan (confidence interval)        | `95`              |
| `estimator`     | Fungsi agregasi (contoh: `mean`, `sum`)           | `np.mean`         |
| `palette`       | Skema warna untuk plot                            | `deep`            |
| `orient`        | Orientasi plot (`"v"` untuk vertikal, `"h"` untuk horizontal) | `v` |
| `order`         | Urutan kategori pada sumbu X                     | `None`            |
| `hue_order`     | Urutan kategori pada parameter `hue`              | `None`            |
| `saturation`    | Intensitas warna bar                              | `0.75`            |
| `dodge`         | Mengatur apakah bar hue akan dipisah atau digabung | `True`            |
| `width`         | Lebar setiap bar                                  | `0.8`             |
| `capsize`       | Panjang garis pada ujung bar (error bar)          | `None`            |
| `errcolor`      | Warna garis error bar                             | `black`           |
| `errwidth`      | Ketebalan garis error bar                         | `None`            |
| `ax`            | Objek matplotlib `axes`                           | `None`            |

##  **4.2. Stripplot**

###  **Pengertian**
`stripplot()` digunakan untuk menampilkan data diskrit atau kategori dalam bentuk titik-titik di sepanjang salah satu sumbu. Plot ini berguna untuk melihat distribusi data dan mengidentifikasi outlier.

Stripplot menampilkan semua data poin dalam dataset tanpa agregasi. Biasanya digunakan untuk dataset yang lebih kecil agar plot tidak terlalu padat.

|---------------|-----------------------------------------------------|---------------|

## 4.3. Swarmplot

`swarmplot()` digunakan untuk membuat scatter plot pada sumbu kategori dengan posisi titik yang disesuaikan agar tidak tumpang tindih. Fungsi ini mirip dengan `stripplot()`, tetapi `swarmplot()` lebih efektif dalam menghindari tumpang tindih antar titik dengan menyesuaikan posisi titik secara otomatis.

### Fungsi `sns.swarmplot()`
```python
sns.swarmplot(x=None, y=None, hue=None, data=None, order=None, hue_order=None, dodge=False, orient=None, color=None, palette=None, size=5, edgecolor='gray', linewidth=0, ax=None, dodge=False)
```

### Parameter Utama:
- **x**: (string, optional) Nama kolom dalam dataset yang akan diplot di sumbu x.
- **y**: (string, optional) Nama kolom dalam dataset yang akan diplot di sumbu y.
- **hue**: (string, optional) Nama kolom untuk membedakan kategori berdasarkan warna.
- **data**: (DataFrame) Dataset yang digunakan untuk membuat plot.
- **order**: (list, optional) Menentukan urutan kategori pada sumbu x atau y.
- **hue_order**: (list, optional) Menentukan urutan kategori warna.
- **dodge**: (bool, default=False) Jika `True`, memisahkan titik-titik berdasarkan kategori dalam `hue`.
- **orient**: (string, optional) Menentukan orientasi plot, bisa berupa 'v' (vertikal) atau 'h' (horizontal).
- **color**: (matplotlib color) Warna untuk titik-titik dalam plot.
- **palette**: (dict atau seaborn color palette) Skema warna yang digunakan.
- **size**: (numeric) Ukuran titik-titik dalam plot.
- **edgecolor**: (matplotlib color) Warna tepi titik.
- **linewidth**: (numeric) Lebar garis tepi titik.
- **ax**: (matplotlib axes) Sumbu pada mana plot akan digambar.

### Contoh Kode:
Menggunakan dataset Titanic untuk membuat plot swarmplot berdasarkan jenis kelamin dan usia:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Dataset Titanic
titanic = sns.load_dataset("titanic")

# Membuat Swarmplot
plt.figure(figsize=(8, 8))
sns.swarmplot(x="sex", y="age", hue="sex", data=titanic)
plt.title("Swarmplot of Age by Sex on Titanic Dataset")
plt.show()
```

### Penjelasan Plot:
- Sumbu **x** menunjukkan kategori jenis kelamin (`sex`).
- Sumbu **y** menunjukkan usia penumpang (`age`).
- Warna dalam plot membedakan kategori berdasarkan jenis kelamin.
- Dari plot ini, kita dapat melihat distribusi usia penumpang berdasarkan jenis kelamin, dengan sebagian besar penumpang berusia antara 18 hingga 40 tahun.

### Tips:
Jika dataset memiliki banyak titik, `swarmplot()` dapat menjadi lambat. Untuk dataset besar, disarankan menggunakan `stripplot()` dengan parameter `jitter` untuk menambahkan noise ke data, yang lebih efisien untuk visualisasi.

---


````python
import seaborn as sns
import matplotlib.pyplot as plt

# Dataset Titanic
titanic = sns.load_dataset("titanic")

# Membuat Swarmplot
plt.figure(figsize=(8, 8))
sns.swarmplot(x="sex", y="age", hue="pclass", data=titanic, palette='bright', dodge=True)
plt.title("Swarmplot of Age by Sex on Titanic Dataset")
plt.show()
````

# 4.4 Boxplot

**Boxplot** adalah cara yang mudah digunakan untuk memvisualisasikan **distribusi data berdasarkan kuartil-kuartilnya**. Boxplot memiliki garis vertikal yang memanjang dari kotak (box) yang disebut dengan **whiskers**. Garis whiskers ini menunjukkan **variabilitas** data di luar kuartil atas dan bawah. Oleh karena itu, boxplot juga dikenal dengan istilah **box-and-whisker plot** atau **diagram kotak dan garis**.  
Setiap **outlier** (data yang berada jauh di luar distribusi normal) akan ditampilkan sebagai **titik-titik individu** di luar boxplot.

##  Contoh Kode:
```python
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat ukuran figure
plt.figure(figsize=(8,8))

# Membuat boxplot
sns.boxplot(x="Survived", y="Age", data=full_data)
```

## 4.5 Violinplot Plot

### **Penjelasan Violinplot**
`violinplot()` adalah kombinasi antara **boxplot** dan **estimasi kepadatan kernel (KDE)**. Plot ini memudahkan kita untuk menganalisis **distribusi data** dengan lebih baik.

- Bagian **kuartil dan whisker dari boxplot** ditampilkan di dalam violin.
- Karena violinplot menggunakan **KDE**, bagian yang lebih lebar dari violin menunjukkan **kepadatan yang lebih tinggi**, sedangkan bagian yang sempit menunjukkan **kepadatan yang lebih rendah**.

---

###  **Contoh Kode Violinplot**
```python
import seaborn as sns
import matplotlib.pyplot as plt

# Membuat violinplot
sns.violinplot(x="Survived", y="Age", hue="Sex", data=full_data)
plt.show()
```

---

###  **Penjelasan Parameter:**
| Parameter      | Deskripsi                                          |
|----------------|----------------------------------------------------|
| `x`            | Kolom yang ditampilkan di sumbu X.                 |
| `y`            | Kolom yang ditampilkan di sumbu Y.                 |
| `hue`          | Warna berdasarkan kategori tertentu (opsional).    |
| `data`         | Dataset yang digunakan untuk membuat violinplot.   |

---

###  **Bagaimana Cara Membaca Violinplot?**
1. **Bagian dalam violin** menunjukkan **boxplot** (Q1, median, Q3).
2. **Lebar violin** menunjukkan **kepadatan data** pada nilai tertentu.
   - Bagian yang lebih lebar = Data lebih banyak/kepadatan lebih tinggi.
   - Bagian yang lebih sempit = Data lebih sedikit/kepadatan lebih rendah.
3. **Hollow dots atau garis** di bagian tengah menunjukkan **median**.

---

# 5. Regression Plots dalam Seaborn

Dalam analisis data, kita sering bekerja dengan dataset yang memiliki banyak variabel kuantitatif. Salah satu tujuan analisis adalah menemukan hubungan antara variabel-variabel tersebut. Plot regresi adalah alat yang sangat berguna untuk memvisualisasikan hubungan ini dengan menambahkan garis regresi ke dalam plot.

Plot regresi juga dapat digunakan untuk memeriksa multikolinearitas, yaitu situasi di mana beberapa variabel independen dalam dataset sangat berkorelasi satu sama lain. Kita perlu mengidentifikasi multikolinearitas ini agar dapat mengambil tindakan yang tepat dalam membangun model regresi.

## 5.1. lmplot

Fungsi `lmplot()` digunakan untuk membuat plot regresi sederhana. Fungsi ini memerlukan parameter `data`, dan variabel `x` serta `y` harus ditentukan dalam bentuk string. Data yang digunakan harus dalam bentuk "long-form", yaitu data di mana setiap kolom mewakili variabel tertentu.

### Contoh Kode:

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.lmplot(x="Age", y="Fare", data=full_data)
plt.show()
```

Pada contoh di atas, kita membuat plot regresi untuk melihat hubungan antara umur (`Age`) dan tarif (`Fare`). Garis regresi akan ditambahkan secara otomatis ke dalam plot.

---

## 5.2. regplot

Fungsi `regplot()` mirip dengan `lmplot()`, tetapi lebih fleksibel karena dapat menerima variabel `x` dan `y` dalam berbagai format, termasuk array numpy, objek Series dari pandas, atau referensi variabel dalam DataFrame pandas.

### Contoh Kode 1:

```python
plt.subplots(figsize=(10, 10))

```

Pada contoh ini, kita membuat plot regresi dengan ukuran plot yang lebih besar menggunakan `plt.subplots(figsize=(10, 10))`.

### Contoh Kode 2:

```python
plt.subplots(figsize=(10, 10))
sns.regplot(x="Fare", y="Fare", data=full_data)
plt.show()
```

Pada contoh kedua, kita membuat plot regresi di mana variabel `x` dan `y` adalah variabel yang sama (`Fare`). Hal ini berguna untuk memeriksa distribusi nilai di sepanjang garis diagonal.

---

## Perbedaan lmplot dan regplot

| **lmplot**                                                                              | **regplot**                         |
| --------------------------------------------------------------------------------------- | ----------------------------------- |
| Menggunakan FacetGrid                                                                   | Menggunakan Axes-level              |
| Memerlukan data dalam format long-form                                                  | Dapat menerima berbagai format data |
| Digunakan untuk plot yang lebih kompleks dengan pembagian subplots berdasarkan kategori | Digunakan untuk plot sederhana      |

Kedua fungsi ini sangat berguna dalam analisis regresi. Pilih fungsi yang sesuai dengan kebutuhan analisis Anda. Jika Anda membutuhkan plot yang lebih kompleks dengan kategori yang berbeda, gunakan `lmplot()`. Jika Anda hanya memerlukan plot sederhana, gunakan `regplot()`.

# 6. Matrix Plots

## 6.1. Heatmap

### Pengertian Heatmap
Heatmap adalah representasi grafis dari data di mana nilai-nilai individu dalam sebuah matriks atau tabel direpresentasikan dengan warna. Heatmap sangat berguna untuk memvisualisasikan data yang kompleks, terutama ketika Anda ingin memahami hubungan atau pola antar variabel dalam dataset.

Heatmap sering digunakan untuk:
- Menunjukkan matriks korelasi antara variabel dalam dataset.
- Mengidentifikasi pola atau tren dalam data yang besar.
- Menemukan nilai ekstrem (maksimum atau minimum) dalam data.
- Membandingkan berbagai fitur dalam dataset dengan mudah.

### Kapan Heatmap Digunakan?
Heatmap digunakan ketika:
- Anda memiliki dataset dengan banyak variabel dan ingin memahami hubungan antar variabel.
- Anda ingin dengan cepat mengidentifikasi korelasi positif atau negatif antara variabel.
- Anda ingin menampilkan data matriks dalam format visual yang mudah dipahami.
- Anda ingin menemukan pola umum dalam data, seperti kelompok yang berkorelasi tinggi atau outlier.

Contoh penggunaan heatmap:
- Dalam analisis keuangan untuk melihat hubungan antara berbagai indikator ekonomi.
- Dalam analisis data kesehatan untuk memeriksa hubungan antara faktor risiko dan hasil kesehatan.
- Dalam data ilmiah untuk menampilkan hasil eksperimen dalam bentuk matriks.

### Cara Membuat Heatmap dengan Seaborn
Berikut adalah langkah-langkah untuk membuat plot matriks korelasi dari data kita menggunakan **heatmap**:

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat heatmap
sns.heatmap(full_data.corr(), cmap="YlGnBu", annot=True, fmt=".2f")

# Menampilkan plot
plt.show()
```

### Penjelasan Sintaks dan Parameter
#### 1. `sns.heatmap()`
Fungsi **`sns.heatmap()`** digunakan untuk membuat heatmap dari data matriks atau dataframe. Berikut adalah beberapa parameter yang sering digunakan:

| Parameter | Deskripsi |
|-----------|-----------|
| `data` | Data yang akan divisualisasikan dalam bentuk matriks atau dataframe. |
| `cmap` | Skema warna yang digunakan dalam heatmap. Contoh: "YlGnBu", "coolwarm", dll. |
| `annot` | Jika `True`, menampilkan nilai dalam setiap sel heatmap. |
| `fmt` | Format nilai anotasi. Contoh: `.2f` untuk menampilkan dua angka desimal. |
| `linewidths` | Ketebalan garis pemisah antar sel dalam heatmap. |
| `linecolor` | Warna garis pemisah antar sel. |



### Interpretasi Heatmap
Dari heatmap ini, kita dapat mencatat bahwa korelasi antar variabel umumnya lemah. Heatmap membantu kita dengan cepat mengidentifikasi hubungan kuat dan lemah antara variabel dalam dataset.

# 7. Multi-Plot Grids

## 7.1. Face Grids

Pendekatan yang berguna untuk mengeksplorasi data berdimensi sedang adalah dengan menggambar beberapa instance dari plot yang sama pada subset yang berbeda dari dataset Anda. Teknik ini biasa disebut sebagai plot "lattice" atau "trellis", dan terkait dengan ide "small multiples". Untuk menggunakan fitur ini, data Anda harus berada dalam DataFrame Pandas.

Facet grid membentuk matriks panel yang didefinisikan oleh baris dan kolom dengan membagi variabel-variabel. Karena adanya panel, satu plot terlihat seperti beberapa plot. Ini sangat membantu untuk menganalisis semua kombinasi dalam dua variabel diskrit.

Objek FacetGrid mengambil DataFrame sebagai input dan nama-nama variabel yang akan membentuk dimensi baris, kolom, atau hue dari grid. Variabel-variabel tersebut sebaiknya bersifat kategorikal, dan data pada setiap level variabel akan digunakan untuk sebuah facet di sepanjang sumbu tersebut.

Keuntungan menggunakan FacetGrid adalah kita dapat memasukkan variabel lain ke dalam plot. Kita dapat membuat banyak facet kolom dan menyelaraskannya dengan baris grid.

### FacetGrid.map
Pendekatan utama untuk memvisualisasikan data pada grid ini adalah dengan metode `FacetGrid.map()`.

```python
# Menginisialisasi objek FacetGrid
g = sns.FacetGrid(full_data, col='Survived', row='Pclass')

# Memetakan histogram pada grid
g.map(plt.hist, 'Age')

# Menambahkan legenda
g.add_legend()

# Menampilkan plot
plt.show()
```



````python
data = sns.load_dataset('titanic')

g = sns.FacetGrid(data, col='survived', row='pclass')

# Memetakan histogram pada grid
g.map(plt.hist, 'age')

# Menambahkan legenda
g.add_legend()

# Menampilkan plot
plt.show()
````
